```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import mss
from PIL import Image
```

```python
game_img = cv2.imread('/content/drive/MyDrive/Code/Knife Hit BOT/game_screen.png', cv2.IMREAD_UNCHANGED)
not_knife_img2 = cv2.imread('/content/drive/MyDrive/Code/Knife Hit BOT/not_knife2.png', cv2.IMREAD_UNCHANGED)
not_knife_img3 = cv2.imread('/content/drive/MyDrive/Code/Knife Hit BOT/not_knife3.png', cv2.IMREAD_UNCHANGED)
knife_img = cv2.imread('/content/drive/MyDrive/Code/Knife Hit BOT/knife.png', cv2.IMREAD_UNCHANGED)

searching = not_knife_img2
```

```python
cv2_imshow(game_img)
cv2_imshow(not_knife_img2)
cv2_imshow(knife_img)
```

```python
result = cv2.matchTemplate(game_img, searching, cv2.TM_CCOEFF_NORMED)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
max_val
```

```python
w = searching.shape[1]
h = searching.shape[0]

cv2.rectangle(game_img, max_loc, (max_loc[0] + w, max_loc[1] + h), (0,255,255), 2)
```

```python
threshold = 0.9
yloc, xloc = np.where(result >= threshold)
len(xloc)
```
